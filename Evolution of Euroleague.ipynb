To analyse team statistics per game, we need to find number of games that every team played per season.
In EuroLeague game length is 40 minutes. There must be 5 players on the field at any moment, therefore there are 200 minutes of game time divided between all team players in one game. Knowing that we can calculate how many games are recorded per team.
Let's see what we get ! 

```python
# Grouping by Season and team
game_time_euro = 200
bySeasonTeam = players_euro.groupby(by=['Season', 'Team']).sum()
bySeasonTeam['GP'] = bySeasonTeam['MIN'] / game_time_euro
bySeasonTeam.head()
sns.scatterplot(data=bySeasonTeam.sum(level=0), x=bySeasonTeam.sum(level=0).index, y='GP')
plt.xticks(rotation=60)
plt.title('Recorded games per season')
plt.show()
```

We can notice there is some data imbalance as later seasons have more documented games. Hopefully this does now skew our upcoming analysis.

Now I will divide seasons into into 4 different eras as I believe they are pretty similar to each other and find per game averages:

```python
bySeasonTeam = bySeasonTeam.reset_index()

era_2000 = bySeasonTeam.loc[(bySeasonTeam['Season'] == '2000') |
                (bySeasonTeam['Season'] == '2001') | (bySeasonTeam['Season'] == '2002') |
                (bySeasonTeam['Season'] == '2003') | (bySeasonTeam['Season'] == '2004')]

era_2005 = bySeasonTeam.loc[(bySeasonTeam['Season'] == '2005') | (bySeasonTeam['Season'] == '2006') |
                (bySeasonTeam['Season'] == '2007') | (bySeasonTeam['Season'] == '2008') | 
                (bySeasonTeam['Season'] == '2009')]

era_2010 = bySeasonTeam.loc[(bySeasonTeam['Season'] == '2010') | (bySeasonTeam['Season'] == '2011') |
                (bySeasonTeam['Season'] == '2012') | (bySeasonTeam['Season'] == '2013') |
                (bySeasonTeam['Season'] == '2014')]

era_2015 = bySeasonTeam.loc[(bySeasonTeam['Season'] == '2015') | (bySeasonTeam['Season'] == '2016') |
                (bySeasonTeam['Season'] == '2017') | (bySeasonTeam['Season'] == '2018') |
                (bySeasonTeam['Season'] == '2019')]

era_2000 = era_2000.groupby(by='Team').sum()
era_2005 = era_2005.groupby(by='Team').sum()
era_2010 = era_2010.groupby(by='Team').sum()
era_2015 = era_2015.groupby(by='Team').sum()
df = pd.concat([era_2000, era_2005, era_2010, era_2015], keys=['2000 - 2004', '2005 - 2009', '2010 - 2014', '2015 - 2019'])
df = df.iloc[:].div(df['GP'], axis=0)
df.index.names = ['Season', 'Team']
```

Adding statistics that are the most popular in the basketball analysis is always a great idea:
- FG%
- 3P%
- 2P%
- FT%

```python
df['FG%'] = df['FGM'] / df['FGA']
df['3P%'] = df['3PM'] / df['3PA']
df['FT%'] = df['FTM'] / df['FTA']

df['2PM'] = df['FGM'] - df['3PM']
df['2PA'] = df['FGA'] - df['3PA']
df['2P%'] = df['2PM'] / df['2PA']
```

Finally, let's look at some graphs. First of all, how has the number of points changed during the years ? 

```python
sns.distplot(df.loc['2000 - 2004']['PTS'], label='2000 - 2004')
sns.distplot(df.loc['2005 - 2009']['PTS'], label='2005 - 2009')
sns.distplot(df.loc['2010 - 2014']['PTS'], label='2010 - 2014')
sns.distplot(df.loc['2015 - 2019']['PTS'], label='2015 - 2019')
plt.legend()
plt.title('Density of points per game')
plt.show()
```

It does not look like more points are being scored. One could even argue, that number of scored points have shrinked by quite a bit. Can we figure out why ?
Maybe there are less field goal attempts, or maybe accuracy of players has dropped for some reason ?

```python
g = sns.FacetGrid(data=df.reset_index(), col='Season')
g.map(sns.scatterplot,'FGA','FG%')
plt.show()
```

Actually, it is quite opposite - we can observe increase in field goals attempted, accuracy also slightly increased. it indicates that game is getting faster and more precise. This trend is especially noticeable in 2015 - 2019 seasons. Most likely it is because of rule changes, that were specifically made to reduce time per posession.
This contradicts our previous findings. Maybe there is some other reason that is more important ?

```python
g = sns.FacetGrid(data=df.reset_index(), col='Season')
g.map(sns.scatterplot,'3PA','3P%')
plt.show()
```

There is an obvious trend of increase in three points attempted per game. Although, it is hard to distinguish if the three point accuracy increased. Let's look at it closer.

```python
sns.distplot(df.loc['2000 - 2004']['3P%'], label='2000 - 2004')
plt.axvline(df.loc['2000 - 2004']['3P%'].median(), color='b', linestyle='dashed', linewidth=2)
sns.distplot(df.loc['2005 - 2009']['3P%'], label='2005 - 2009')
sns.distplot(df.loc['2010 - 2014']['3P%'], label='2010 - 2014')
sns.distplot(df.loc['2015 - 2019']['3P%'], label='2015 - 2019')
plt.axvline(df.loc['2015 - 2019']['3P%'].median(), color='r', linestyle='dashed', linewidth=2)
plt.legend()
plt.title('Density of 3P% per game')
plt.show()
```

If we look at the median values, it is apparent that three point accuracy has also slightly increased. It seems that nowadays players throw more threes and they also do it better. Interesting.

Let's give it another shot. Then it would be rational to think that two pointers were way more important in the beggining of 2000s.

Let's analyse other aspects of the game. Has basketball in Europe become more physical over the years ? :

Okay, in early 2000s game was way more physical. It looks like players were getting a lot more fouls than nowadays. More fouls lead to more free throws right ? 

```python
g = sns.FacetGrid(data=df.reset_index(), col='Season')
g.map(sns.scatterplot,'FTA','FT%')
plt.show()
```

Number of free throws attempted clearly decreased in the last 20 years while the accuracy increased by a margin. Way more successful free throws could be the lead reason as of why teams do not score as much. Add more two pointers made on top of that and you get a reasonable gap of scored points.

```python
sns.distplot(df.loc['2000 - 2004']['BLK'], label='2000 - 2004')
sns.distplot(df.loc['2005 - 2009']['BLK'], label='2005 - 2009')
sns.distplot(df.loc['2010 - 2014']['BLK'], label='2010 - 2014')
sns.distplot(df.loc['2015 - 2019']['BLK'], label='2015 - 2019')
plt.legend()
plt.title('Density of blocks')
plt.show()
```

Unexpectedly, the least amount of blocks were made between 2000 - 2004. It looks like teams are getting better not only at offense but defense as well.

# Conclusions
   ## Offense
   Offense has changed a lot over the last 20 years. Game became quicker, with more field goal attempts per game, as there were some rule changes. Nowadays teams give more attention to play around perimeter which leads to more spacing on the court. More space helps to be more precise with the ball, perform better passes and do less mistakes overall. Moreover, players became better at shooting in all categories (2P, 3P, FT) with the most obvious increase in free throws. Three points also improved considerably, despite the fact that three point line was pushed half of meter away from the basket.

   ## Defense
   In this analysis I paid the most attention to offense, but one could say that increase in free throws successes made teams value fouls. As players got better at converting nobody wanted to give easy points to opponent while in early 2000s there was a decent chance fouling team will get away unpunished.

## Overall
All in all, european basketball is definitely on the right path with almost every statistic being better nowadays. The question is - what is the limit for improvement ?
